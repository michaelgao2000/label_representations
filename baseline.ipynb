## Step 2. Figures

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
data_dir = root + '/label_representations/outputs/log/'
```

```python
# read in experiment data
experiments = ["category","composite", "shuffle", "speech", "uniform"]
test_acc = pd.DataFrame(columns=["Label", "Test Accuracy"])

ax = plt.gca()
for exp in experiments:
  train = pd.read_csv(data_dir + "cifar10_" + exp + "_train_log.csv", header=0)
  test = pd.read_csv(data_dir + "cifar10_" + exp + "_log.csv", header=0)

  test_acc.loc[len(test_acc.index)] = [exp, test["Test Acc"][0]]
  color = next(ax._get_lines.prop_cycler)['color']

  plt.plot(train["Epoch"], train["Train Loss"], color=color)
  plt.plot(train["Epoch"], train["Valid Loss"], color=color, linestyle='dashed')

  plt.xlabel("Epoch")
  plt.ylabel("Loss")

  if exp == "category":
    plt.legend(["category"])
    plt.show()
    ax = plt.gca()
handles = []
line1 = Line2D([0], [0], label='composite', color='C0')
line2 = Line2D([0], [0], label='shuffle', color='C1')
line3 = Line2D([0], [0], label='speech', color='C2')
line4 = Line2D([0], [0], label='uniform', color='C3')
handles.extend([line1, line2, line3, line4])

plt.legend(handles=handles)
plt.show()
```
